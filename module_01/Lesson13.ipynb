```python
#Reading and writing files
# Let's write a simple note
# "w" means Write (Overwrites existing file!)
with open("my_notes.txt", "w") as file:
    file.write("This is my first saved file!\n")
    file.write("I can write mulitple lines.\n")

print("File saved successfully.")
```

```python
# Reading from the a file
# "r" means Read
with open("my_notes.txt", "r") as file:
    # .read() pulls the whole file into a string variable
    content = file.read()

print("---Here is what was in the file---")
print(content)
```
