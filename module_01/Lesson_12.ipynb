```python
# Dictionaries in Python

#Creating a Dictionary
# A simple phonebook Dictionary
# Key(Name) : Value(Anything like = "strings", Numbers, Booleans)
phonebook = {
    "Prateek" : 4679876454635,
    "Alice" : 998746564545,
    "Bob" : 8784434354
}

# Accessing the data from the dictionaries
print(phonebook["Prateek"]) # output = 4679876454635
# print(phonebook[]) # -> Error = SyntaxError: invalid syntax. Perhaps you forgot a comma?
# print(phonebook) # -> passes whole dictionary to the reciver 
print(phonebook["Bob"])

#Adding new items to the Dictionary
phonebook["Davide"] = 24585465654 # added new item to the phonebook Dictionary
print(phonebook)

#Changing value in the dectionary
phonebook["Alice"] = 65654464 # the "Alice" key value is changed
print(phonebook)
```
