```python
# Functions in python


# . How to Define Your First Function
# Let's make a simple function. The syntax has 4 parts:

# The def keyword (short for "define").

# The function's name (use snake_case, just like variables).

# Parentheses () after the name.

# A colon : at the end.

# All the code inside the function must be indented.
```

```python
# This is the DEFINITION (writing the recipe)

# This code does NOT run yet. It's just being stored in memory

def greet_user():
    print("Hello, Gemini")
    print("Welcome to Milstrone 3")
    print("Let's learn about functions.")

# This is the CALL (using the recipe)
# Now we "call" the function by its name. This is where the code runs.

print("----Running the function the first time----")
greet_user() # function call

print("----Running it againe---")
greet_user() # agine function call
```

```python
# parameters in functions

# 'name' is a parameter (an "ingredient" placeholder)

def greet_by_name (name) :
    print(f"Hello, {name}!")
    print(f"Welcome to programming world")


# greet_by_name(input("Enter the name")) # what we are inputing is called argument

# "Alice" is an ARGUMENT (the "real" ingredient)
greet_by_name("Alice")

# "Bob" is an ARGUMENT
greet_by_name("Bob")
```

```python
# RETURN in functions

# This function's job is to add two numbers and RETURN the result

def add_numbers(num1, num2):
    total = num1 + num2
    return total # it gives back the sumvalue of num1 and num2

# --------CALLING IT--------

# We "call" the function and "catch" the result it hands back
# The value from 'return' gets stored in our "sum_result" variable

sum_result = add_numbers(10, 48) # giving two arguments to the add_numbers() function seperated by and returned value stored in the sum_result variable ,
print(sum_result)
print(f"can be used for other calculation or usage : {sum_result + 25}") 
```

```python
# the TO-DO-list and password generator combined


import random
import string

# ---------Password generator function-----------
def password_generator():
    # Charecter Pools 
    # letters
    letters = list(string.ascii_lowercase) # all lower case letter are added to the list

    # numbers
    numbers = list(string.digits)# all the digits are added to the list

    #symbols
    symbols = list(string.punctuation) # most of the symbols are added to the list

    # input for letters count
    letters_count = int(input("How many letters you want to add to the password"))

    # input for numbers count
    numbers_count = int(input("How many numbers you want to add to the password"))

    # input for symbols count
    symbols_count = int(input("How many symbols you want to add to the password"))


    #empty  password list
    password_list = []

    # for i in range(letters_count):
    #     password_list.append(random.choice(letters))

    # for i in range(numbers_count):
    #     password_list.append(random.choice(numbers))

    # for i in range(symbols_count):
    #     password_list.append(random.choice(symbols))


    # updated code using function
    def charecter_count(count, type):

        for i in range(count):
            password_list.append(random.choice(type))


    charecter_count(letters_count, letters)
    charecter_count(numbers_count, numbers)
    charecter_count(symbols_count, symbols)


    random.shuffle(password_list)

    final_password_string = "".join(password_list) # This tells Python, "Take an empty string "" and use it to join every single item in password_list."

    # for i in password_list :

    #     # converting list to string
    #     final_password_string += i

    print(f"Your strong password is : '{final_password_string}'\nThank you!")


# -----------TO-DO list function-----------

def to_do_list_fun():


    # for spacing and gaps
    def for_spacing_purpose(msg):
        # for some spacing between the updated task and privious task
            print("                               ")
            print(msg)
            print("                               ")

    # empty list for collection of tasks
    tasks_to_do = []


    # choice = input("For adding task enter : 'add'\nFor checking the tasks enter : 'show'\nFor For completed tasks enter task number\nFor quiting the menu enter 'quit'" )

    # forever loop 
    while True:

        choice = input("For adding task enter : 'add'\nFor checking the tasks enter : 'show'\nFor For completed tasks enter : 'done'\nFor quiting the menu enter : 'quit'" )
    
        if choice == "add" :
        # task adding to the list
            tasks_to_do.append(input("Enter your task to add"))
            continue

        elif choice == "show":
        
            if len(tasks_to_do) == 0:
                print(f"There are no tasks to complete, so enter the new tasks")
            else:
                print("---Your tasks---")
                for index, items in enumerate(tasks_to_do):
                    # item number with serial numbers
                    print(f"{index + 1}. {items}")
            continue

        elif choice == "done":

            choice_number = int(input("Enter the serial number of the task completed"))

            if choice_number == 0 :
            # if the value is 0 then there are no tasks to remove
                print(f"There is no completed tasks to remove form the list")

            elif choice_number <= len(tasks_to_do):
            # the serial number starts from 1 so to remove the mentioned number
                tasks_to_do.pop((choice_number) - 1)

                # calling the function for spacing
                for_spacing_purpose("task list Updating......")

                # if there are no items to show after removing
                if len(tasks_to_do) == 0 :
                    print(f"All tasks are completed, Add new tasks to the list")
                    continue
                
                # calling the function for spacing
                for_spacing_purpose("Updated task list ")

                
                # for loop to run through all items
                for index, items in enumerate(tasks_to_do):
                # updated tasks list
                    print(f"{index + 1}. {items}")

            continue

        # exiting the while loop
        elif choice == "quit":
            # calling the function for spacing
            for_spacing_purpose(" ")
            print(f"Goodbye!")
            break

# main_menu function for calling different functions again and again

while True :
    choice = input(f"Press 1 for Password generator, 2 for TO-DO-List, or type 'quit' to exit")
    
    if choice == "1":
        password_generator()

    elif choice == "2":
        to_do_list_fun()

    elif choice == "quit":
        break


```

```python
# Matric converter

def celcius_to_fahrenheit(celsius):

    # Fahrenheit value
    fahrenheit = (celsius * (9/5)) + 32
    # converted value return 
    return fahrenheit

def kilograms_to_Pounds(kg):

    #pounds convertion
    pounds = kg * 2.20462

    # Return to the called value
    return pounds

def kilometers_to_miles(km):

    # miles convertion
    miles = km * 0.621371
    
    # Return to the called value
    return miles





while True :

    # Choice selection
    choice = input("Enter '1' to convert Celsius to Fahrenheit\n'2' to Convert Kilograms to Pounds\n'3' to Convert Kilometers to Miles\nType 'quit' to exit")
    
    if choice == "1":

        # users temp in C
        temp_input_c = float(input("Enter the tempr you want to convert"))

        # converted tempr recieved from celcius_to_fahrenheit function
        converted_temp = round(celcius_to_fahrenheit(temp_input_c), 2)

        # output tempr
        print(f"The converted tempreture from {temp_input_c}C to fahrenheit is : {converted_temp}F")

        #continue -> not requied here

    elif choice == "2":

        #users weight in kg
        weight_input_kg = float(input("Enter the weight in kg to convert it to pounds"))

        # converted value
        converted_weight = round(kilograms_to_Pounds(weight_input_kg), 2)

        # output weight
        print(f"The converted value of {weight_input_kg}kg to pounds is : {converted_weight}pound")

        #continue

    elif choice == "3" :

        #user input distance
        distance_input_km = float(input("Enter the distance in km to convert it to miles"))

        #converted value
        converted_distance = round(kilometers_to_miles(distance_input_km), 2)

        # output
        print(f'The converted value of {distance_input_km}km to miles is : {converted_distance}mile')

        #continue

    elif choice == "quit":

        #last msg

        print("Thank you for your usage :)\nGOODBYE")

        break
```
