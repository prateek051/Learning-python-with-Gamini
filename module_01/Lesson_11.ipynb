```python
#Object Oriented programing
# classes in python

#this is the blueprint for a Dog
# Right now it's empty.
class Dog:
    pass # 'pass' is a placeholder that means 'do nothing'

#use the class name with () to build an object
my_first_dog = Dog()
my_second_dog = Dog()

print(type(my_first_dog))
print(my_second_dog)
```

```python
class Dog:
    #This is the CONSTRUCTOR
    #it runs when we do Dog("Fido", 5)
    def __init__(self, name_param, age_param):
        # 'self' refers to the new object being created

        #we are "attaching" the data to the object
        self.name = name_param
        self.age = age_param

    # Method (a skill)
    def bark(self):
        # 'self.name' lets the dog use its *own* name
        print(f"{self.name} says: woof!")

    def get_human_age(self):
        human_years = self.age * 7
        print(f"{self.name} is {human_years} in human years.")


# ----1. Let's build two dog from out blueprint
dog_1 = Dog("Fido", 5)
dog_2 = Dog("Spot", 2)
dog_3 = Dog("julo", 3)

# ----2. Let's check their data----
# We use the dot . to access their internal data
print(f"Dog 1's name is : {dog_1.name}")
print(f"Dog 2's name is : {dog_2.name}")

# ----3. Let's use their skills (methods)
# we use the dot . to call their methods

dog_1.bark() #first method
dog_2.bark() #first method

dog_1.get_human_age()# second method
dog_2.get_human_age()# second method

dog_3.get_human_age() # second method
```

```python
# practice  assignement

# Student class
class Student:

    #Constructor
    def  __init__(self, name_param, id_param):
        self.name  = name_param
        self.id = id_param
        
    
    def introduce(self):
        print(f"Hello, My name is {self.name} and my student ID is {self.id}.")


# created objedt instances
student_1 = Student("Arun", 255)

# second object instances
student_2 = Student("Alice", 258)

# method using from "Student" Class
student_1.introduce()
student_2.introduce()
```

```python
# TO - DO -List revised using OPP

# to-do-list class
class To_do_list:

    # constructor function
    def __init__(self):
        # Empty list
        self.tasks = []

    # Tasks adding to the "tasks " list
    def add_task(self, task_name):
        # appending the tasks
        self.tasks.append(task_name)

        # confirmation of the task added
        print(f"Added task : {task_name}")


    def show_task(self):
        
        # task display
        for index , item in enumerate(self.tasks):

            # condittion check for empty list
            if len(self.tasks) == 0:
                print("Empty")
                
            else:
                print(f"{index + 1}, {item}")


my_list_1 = To_do_list()

# my_list_1.add_task("wake up early")
my_list_1.show_task()

```
