```python
#Object Oriented programing
# classes in python

#this is the blueprint for a Dog
# Right now it's empty.
class Dog:
    pass # 'pass' is a placeholder that means 'do nothing'

#use the class name with () to build an object
my_first_dog = Dog()
my_second_dog = Dog()

print(type(my_first_dog))
print(my_second_dog)
```

```python
class Dog:
    #This is the CONSTRUCTOR
    #it runs when we do Dog("Fido", 5)
    def __init__(self, name_param, age_param):
        # 'self' refers to the new object being created

        #we are "attaching" the data to the object
        self.name = name_param
        self.age = age_param

    # Method (a skill)
    def bark(self):
        # 'self.name' lets the dog use its *own* name
        print(f"{self.name} says: woof!")

    def get_human_age(self):
        human_years = self.age * 7
        print(f"{self.name} is {human_years} in human years.")


# ----1. Let's build two dog from out blueprint
dog_1 = Dog("Fido", 5)
dog_2 = Dog("Spot", 2)
dog_3 = Dog("julo", 3)

# ----2. Let's check their data----
# We use the dot . to access their internal data
print(f"Dog 1's name is : {dog_1.name}")
print(f"Dog 2's name is : {dog_2.name}")

# ----3. Let's use their skills (methods)
# we use the dot . to call their methods

dog_1.bark() #first method
dog_2.bark() #first method

dog_1.get_human_age()# second method
dog_2.get_human_age()# second method

dog_3.get_human_age() # second method
```
