```python
# The Wearhouse Manager
class inventoryManager :
    # To store data in the WearHouse
    def __init__(self):
        self.inventory = {}

        #Load data method call
        self.load_data()


# adding the item to the wearhouse
    def add_item(self, name, quantity) :
        #checking for existing item
        if name in self.inventory :
            #changing the quantity of the items
            self.inventory[name] += quantity
            # print( f"{name} = {self.inventory[name]}")

            #if new item is given then adding to the wearhouse
        else :
            self.inventory[name] = quantity

        self.save_data()


    #Removing the items from the inventory
    def remove_item(self, name, quantity) :
        #Removing the item from the wearhouse
        if name in self.inventory:
             # check for item avalability
            if self.inventory[name] >= quantity :
                 #substract the quantity from the available
                 self.inventory[name] -=quantity
            else:
                print(f"You don't have enough quantity in your WearHouse")
        else:
            print("You don't have This item in your wearhouse")
        
        # Calling save_data method
        self.save_data()


    # Inventory display
    def show_inventory(self) :
        print(f"---Items available in inventory---")
        # inventory output
        for items in self.inventory:
            #printing the items
            print(f"{items} : {self.inventory[items]}")


    #Save inventory data
    def save_data(self) :
        # creating the new inventory file and adding the items to the file
        with open("inventory.txt", "w") as file :
            # looping through all the items
            for items in self.inventory :
                file.write(f"{items} : {self.inventory[items]}\n")

    # Data loading from text file for every time it runs
    def  load_data(self) :
        try :
            # Try to open the file
            with open("inventory.txt", "r") as file :
                for line in file :
                    # clean the line (remove \n)
                    clean_line = line.strip()
                    #split "apple : 50 " into ["Apple", "50"]
                    name , count = clean_line.split(" : ")
                    # add back to the dectionary
                    self.inventory[name] = int(count)

            print("Data loaded successfully!")
            print()
        except FileNotFoundError :
            # If file doesn't exist yet (first run), just do nothing
            print("No saved file found. Starting fresh.")

wearhouse_1 = inventoryManager()
        
while True :
    # user choice of operation
    choice = input("Enter '1' for  Add strock\n'2' for Remove Stock\n'3' for View Inventory\n'4' to Quit")

    # choice check
    if choice == "1" :
        item_name = input("Enter the item name to add")
        quantity_count = int(input("Enter the quantity of the item"))
        wearhouse_1.add_item(item_name, quantity_count)
    
    elif choice == "2" : 
        item_name = input("Enter the item name to remove")
        quantity_count = int(input("Enter the quantiy of the item to remove"))
        wearhouse_1.remove_item(item_name, quantity_count)

    elif choice == "3" :
        wearhouse_1.show_inventory()

    elif choice == "4":
        break


```
