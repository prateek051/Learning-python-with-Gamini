```python
#Datatypes in python

#A box holding a NUMBER
box_1 = 10

# A box holding TEXT(a string)
box_2 = "10"

print(type(box_1)) # Number
print(type(box_2)) # string
```

```python
# adding NUMBER + NUMBER
print(box_1 + 5)

# print(box_2 + 5) # TypeError -> NUMBER and STRING cannot be added
```

```python
# String addition (concatenation)

first_name = "Prateek"
last_name = "Sharma"

full_name = first_name + " " + last_name # concatenation
print(full_name)
```

```python
# Arithemetic operators
# + (Addition)
# - (Subtraction)
# * (Multiplication)
# / (Division)

# 1. Store your numbers in variables
pizz_cost = 15
pizzas_needed = 4
drink_cost = 2.50
drink_needed = 8

#2. Do the math
total_pizza_cost = pizz_cost * pizzas_needed
total_drink_cost = drink_cost * drink_needed

#3. Get the final total
total_bill = total_pizza_cost + total_drink_cost

#4. Print the result
print(f"The total cost for the pizza is {total_pizza_cost}")
print(f"The total cost for the drinks {total_drink_cost}")

print("--------------")
print(f"Your grand total is : {total_bill}")
```
