```python
# input function
# input() -> asks for input from user

#1. Ask the user for their name and wait for an answer
# The answer gets stored in the "user_name" box
user_name = input("Hello, I'm the intern. What is your name?\n")

# printing the user name 

print(f"Wow, it's a pleasure to meet you, {user_name}")

# !input function always returns "string" type 
```

```python
# orofit calculator using user input

revenue_amount = float(input("Enter your total revenue\n"))

cost_amount = float(input("Enter your total cost\n"))

profit = revenue_amount - cost_amount

print(f"Your total revenue amount is {revenue_amount} and your total cost is {cost_amount} so your total profit is {profit}")
```
