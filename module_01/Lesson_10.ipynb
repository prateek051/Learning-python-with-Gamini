```python
# Recursion in python

def countdown_recursive(n):
    #1. The base case (The "Stop condition")
    # This is the most important part

    if n == 0:
        print("Blast off!")

    #2. The RECURSIVE STEP (call itself)
    else:
        print(n)
        #Call the *same function* but with a smaller problem (n-1)

        countdown_recursive(n-1)

print("------starting recursive countdown-----")
countdown_recursive(3)

```

```python
# factorial function
def factorial(n):

    # Base case
    if n == 1 and n == 0:
        return 1
    # Recursive step
    else:

        return n * factorial(n-1)
    
# user input
user_input = int(input("Enter a number to get its factorial value"))

# function return catch
factorial_return = factorial(user_input)

print(factorial_return)

```
