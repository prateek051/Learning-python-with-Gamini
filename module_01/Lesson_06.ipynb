```python
#Loops in python
# for loop
# 1. Out list
guest_list = ["Alice", "Bob", "Diana"]



# 2. Out loop
# We;ll invent the temp variable "guest_name"
print("---Sending Invitations---")

for guest_name in guest_list :
    # This indented code runs 3 times
    print(f"Sending an invitation to : {guest_name}")

print("---All invitations sent!---")

```

```python
#Lesson 6 practice assignment
#creat a list of numbers
general_numbers = [10, 34, 20, 59, -332, -4832, -56, 22, 54, -32 , 0, 43, -654, 6589]

positive_numbers = []
negative_numbers = []


for num in general_numbers :
    if num > 0 :
        positive_numbers.append(num)

    elif num == 0 :
        continue

    else:
        negative_numbers.append(num)

print(f"if zero is present it is not stored in the lists")
# positive number list
print(f"The possitive numbers in general numbers are : {positive_numbers}")

# negetive number list
print(f"The negetive numbers in general numbers are : {negative_numbers}")
```
