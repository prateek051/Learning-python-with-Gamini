```python
# Smart Expenses Tracker

# Class transaction to keep the data
class Transaction :
    def __init__(self, date, category, amount) :
        self.date = date
        self.category = category
        self.amount = amount

class ExpenseManager :
    def __init__(self) :
        #We don't store numbers here, We store transaction OBJECTS
        self.transactions = []

    def add_expense(self, date, category, amount) :
        # 1. Create the Recipt object
        new_receipt = Transaction(date, category, amount)

        # 2. Store the Receipt object in the list
        self.transactions.append(new_receipt)


    def get_total_by_category(self, category_to_check) :
        total = 0

        # Loop through every receipt object in out list
        for t in self.transactions :
            # check the category attribute of that specific object
            if t.category == category_to_check :
                total += t.amount

        return total
    
expense_1 = ExpenseManager()
expense_1.add_expense("20-2-2025", "Food", 50)
expense_1.add_expense("31-4-2024", "Food", 350)
expense_1.get_total_by_category("Food")

```
