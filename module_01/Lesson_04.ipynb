```python
# Conditional statements
# == (Double equals): "Is this exactly equal to that?"

# age == 18 (Asks: "Is the value in age exactly 18?")

# Warning! A common mistake is using a single =, which is for assignment (putting things in boxes). == is for comparison (asking a question).

# != (Not equals): "Is this not equal to that?"

# > (Greater than): "Is this number bigger than that one?"

# < (Less than): "Is this number smaller than that one?"

# >= (Greater than or equal to)

# <= (Less than or equal to)

age = 20
print(age > 18) # Python will print True
print(age == 18)  # Python will print False
```

```python
# only if statement is used in below code
# 1. Ask the user for their age.
age = int(input("Welcome! How old are you? \n"))

#2. Check the condition (The fork in the road)
if age >= 21:
    #This "idented block" only runs if the answer is True.
    print(f"Come on in! You're old enough.:)")

#3. This line is NOT indented, so it runs *after* the if-check is over no mater what.
print(f"Have a nice night! :)")
```

```python
# if-else statement is used in below code
# 1. Ask the user for their age.
age = int(input("Welcome! How old are you? "))

#2. Check the condition (The fork in the road)
if age >= 21:
    #This "idented block" only runs if the answer is True.
    # Door A
    print(f"Come on in! you're old enough. :)")

else: 
    # Door B Block (also indented)
    print(f"Sorry, You are not old enough to enter.")

#3. This line is NOT indented, so it runs *after* the if-check is over no mater what.
print(f"Have a nice night! :)")
```

```python
# lesson-04 assignment

# revenue input
revenue = float(input("Enter your total revenue : "))

#cost input
cost = float(input("Enter your total cost : "))

# profit or loss calculation
profit_or_loss = round((revenue - cost), 2) # round(number, decimals) => it controls the number of decimals to be shown

if profit_or_loss > 0 :
    # profit output
    print(f"Your total revenue is : {revenue}, and Your total cost is : {cost}, so the total profit is : {profit_or_loss}")
    print(f"Congratulations! on your profit")

elif profit_or_loss == 0 :
    # no profit and no loss
    print(f"Your total revenue is : {revenue}, and Your total cost is : {cost}, so the total profit is : {profit_or_loss}")
    print(f"Your total revenue and total cost are equal so no profit no loss!")

else :
    #Loss
    print(f"Your total revenue is : {revenue}, and Your total cost is : {cost}, so the total loss is : {abs(profit_or_loss)}") # abs() -> which is absolute value
    print(f"You made a loss! BETTER LUCK NEXT TIME")

print(f"Thank you for your time with us!\nHave nice day :)")
```
